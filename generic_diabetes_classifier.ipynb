```python
#read input csv data
data_file = pd.read_csv('pima-indians-diabetes.csv')
```

```python
#User input for which column is label for classification model
try_again = 'y'
while(try_again.lower() == 'y'):
    label = input("Enter label name which should match to one of your column name from input csv")
    print("\nLabel entered: "+ label)

    found = False
    for x in data_file.columns:
        if x.lower()==label.lower(): 
            label = x 
            found = True
    if(not found):
        try_again = input("No column found with such name, try again?(y/n)")
    else:
        try_again = 'n'
        print("label '{}'' found".format(label))
    
```

```python
#data without label to be used for train and testing
data_file_without_label = data_file.drop(label, axis=1)
```

```python
data_file_without_label.head()
```

```python
#finding columns which has int64 or float64 datatypes
numeric_cols = (data_file.dtypes== 'float64') | (data_file.dtypes== 'int64') & (data_file.columns != 'Age') & (data_file.columns != label)
```

```python
#all non-numeric datatypes for creating categorical cols features
non_numeric_cols = (data_file.dtypes!= 'float64') & (data_file.dtypes!= 'int64') & (data_file.columns != 'Age') & (data_file.columns != label)
```

```python
non_numeric_cols
```

```python
non_numeric_cols = data_file.columns[non_numeric_cols]
```

```python
data_file[non_numeric_cols].drop_duplicates().values
```

```python
numeric_cols = data_file.columns[numeric_cols]
```

```python
data_file[numeric_cols] = data_file[numeric_cols].apply(lambda x: (x-x.min())/(x.max()-x.min()))
```

```python
data_file
```

```python
y_label = data_file.loc[:,label]
```

```python
y_label
```

```python
x_data = data_file.drop(label,axis=1)
```

```python
x_data
```

```python
variable = {}
```

```python
#check if there is an Age column and make buckets for age groups
if 'Age' in data_file.columns:
    print('yes')
    age = tf.feature_column.numeric_column('Age')
    age_buckets  =tf.feature_column.bucketized_column(age, boundaries=[20,30,40,50,60,70,80])
    variable['Age'] = age_buckets
```

```python
#create features
for x in numeric_cols:
    variable[x] = tf.feature_column.numeric_column(x)
for x in non_numeric_cols:
    variable[x] = tf.feature_column.categorical_column_with_hash_bucket(x,hash_bucket_size=len(data_file[x].drop_duplicates().values))
```

```python
variable.keys()
```

```python
#create train and test data
x_train, x_eval, y_train, y_eval = train_test_split(x_data, y_label, test_size=0.3,random_state=101)
```

```python
#creating Linear Classifier model
model = tf.estimator.LinearClassifier(feature_columns=[variable[x] for x in variable.keys()], n_classes=len(data_file[label].drop_duplicates()))
```

```python
input_func = tf.estimator.inputs.pandas_input_fn(x=x_train,y=y_train,batch_size=80,num_epochs=1000,shuffle=True)
```

```python
train_func = tf.estimator.inputs.pandas_input_fn(x=x_train, y=y_train,batch_size=80,num_epochs=1000,shuffle=False)
```

```python
pred_func = tf.estimator.inputs.pandas_input_fn(x=x_eval,batch_size=8,num_epochs=1,shuffle=False)
```

```python
#training model
model.train(input_fn=input_func, steps=1000)
```

```python
#evaluating model against train data
model.evaluate(input_fn=train_func, steps=1000)
```

```python
#predicting label values using test data
predictions = model.predict(input_fn=pred_func)
```

```python
predictions_records = list(predictions)
```

```python
predictions_records
```

```python
predicted_classes = [x['class_ids'][0] for x in predictions_records]
```

```python
#create a method to test user input
def predict_user_input(model):
    feature={}
    for x in numeric_cols:
        feature[x]=[float(input(f"\nEnter the normalized numerical value for feature '{x}' between 0 and 1"))]
    for x in non_numeric_cols:
        feature[x]=[(input(f"\nEnter the value for feature '{x}'"))]
    feature['Age']=[float(input(f"\nEnter the value for age '{x}' between 0 and 100"))]
    user_input_function = tf.estimator.inputs.pandas_input_fn(x=pd.DataFrame(feature),batch_size=1,num_epochs=1,shuffle=False)
    return list(model.predict(user_input_function))
        
```

```python
#input from user and predicted result with probability
user_result = predict_user_input(model)
result_class = user_result[0]['class_ids'][0]
print(f"\n The predicted class is {result_class} with probability of {round(user_result[0]['probabilities'][result_class]*100,2)}%")
```
